```python
# 1 Basic Aggregation:
import pandas as pd
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
sum=df.groupby('Team')['Salary'].sum()
display(sum)
```

```python
# 2 Multiple Aggregations:
import numpy as np
salary_grp=df.groupby('Position')
salary_grp['Salary'].agg([np.sum, np.mean])
```

```python
# 3 Custom Aggregation:
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df=df.drop(457)
def convert_height(height):
    feet,inches=map(int, height.split('-'))
    return feet+inches/12
df['converted height']=df['Height'].apply(convert_height)
team_grp=df.groupby('Team')
diff=lambda x: (x.max() - x.min())
df['height diff']=team_grp['converted height'].transform(diff)
display(df)
```

```python
# 4 Group-wise Transformation:
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
position_grp=df.groupby('Position')
normalize=lambda x : (x - x.mean()) / x.std()
df['normalized_salary']=position_grp['Salary'].transform(normalize)
display(df)
```

```python
# 5 Filter Groups Based on Aggregated Condition:
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
team_grp=df.groupby('Team')
team_grp.filter(lambda x : x['Salary'].sum()< 10000000) 
```

```python
# 6 Filter Based on Group Size:
team_grp.filter(lambda x: len(x)<3)
```

```python
# 7 Filter Based on Group Mean:
position_grp=df.groupby('Position')
position_grp.filter(lambda x : x['Weight'].mean()<=220)
```
