```python
# 1 str.slice() for column manipulation
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df['Team_Abbreviation']=df['Team'].str.slice(0,3)
display(df)
```

```python
# 2 extracting part of a string using str.slice()
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df['position_short']=df['Position'].str.slice(-2,)
display(df)
```

```python
# 3 function to calculate age in months
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
def age_to_months(age):
    return age*12
df["age_in_months"]=df['Age'].apply(age_to_months)
display(df)
```

```python
# 4 using apply() to convert heights
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df=df.drop(457)
def convert_height(height):
    feet,inches=map(int, height.split('-'))
    return feet+inches/12
df['converted height']=df['Height'].apply(convert_height)
display(df)
```

```python
# 5 Row-wise Application of a Custom Function
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
def weight_and_age(row):
    if row['Weight'] > 200 and row['Age'] > 25:
        return True
    else:
        return False
df['heavy_and_old']=df.apply(weight_and_age,axis=1)
display(df)
```

```python
# 6 using df.aggregate func 
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
age_salary=df[['Age','Salary']]
age_salary.aggregate(['mean','median','std'])
```

```python
# 7 using mean() to find average salary
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
average_salary=df['Salary'].mean()
print(f'average alary: {average_salary}')
```

```python
# 8 Filtering Data Using a Custom Function with apply()
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
def avg_salary(salary):
    if salary>average_salary:
        return True
    else:
        return False
df['above avg salary']=df['Salary'].apply(avg_salary)
display(df)
```

```python
# 9 Using aggregate() with a Lambda Function
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
range=df['Weight'].aggregate(lambda x:x.max()-x.min())
print(f'range of the weight: {range}')
```
