```python
# 1 adding new column using df.assign()
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df=df.drop(457)
def convert_height(height):
    feet,inches=map(int, height.split('-'))
    return feet* .3048 + inches*.0254
df=df.assign(height_in_meter=df['Height'].apply(convert_height))
df=df.assign(BMI=lambda x:x['Weight']/ (x['height_in_meter']**2))
display(df)
```

```python
# 2 Adding a Column Based on Conditional Logic Using df.assign()
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df=df.assign(salary_tier=lambda x:['high' if Salary>5000000 else 'low' for Salary in x['Salary'] ])
display(df)
```

```python
# 3 Selecting Rows and Columns Using df.loc()
import pandas as pd 
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
pg_players=df.loc[df['Position'] == 'PG', :]
display(pg_players)
```

```python
# 4 Modifying Data Using df.loc()
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df.loc[df['College'] == 'Texas', 'Team'] = 'New Team'
display(df)
```

```python
# 5 Selecting Rows and Columns Using df.iloc()
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
df.iloc[:5, :3]
```

```python
# 6 Modifying Specific Cells Using df.iloc()
df=pd.read_csv("C:/Users/anant/Downloads/task/nba (1).csv")
Salary_Column_Index=df.columns.get_loc('Salary')
df.iloc[2, Salary_Column_Index] = 10000000 
display(df)
```
